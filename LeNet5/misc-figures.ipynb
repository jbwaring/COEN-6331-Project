```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from torchvision.datasets import MNIST
from torchvision.transforms import ToTensor
from matplotlib import rcParams

# Set the font to Times New Roman
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['font.size'] = 16
# Load the MNIST dataset
mnist = MNIST(root='./data', train=True, download=True, transform=ToTensor())
x_train, y_train = mnist.data.numpy(), mnist.targets.numpy()

# Select the first digit of each class (0-9)
digits = []
for i in range(10):
    digit = x_train[np.where(y_train == i)[0][0]]
    digits.append(digit)

# Create a figure with a 2x5 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(10, 5))

# Display each digit in its own subplot
for i, ax in enumerate(axs.flat):
    ax.imshow(digits[i].squeeze(), cmap='bwr')
    ax.set_title(f"Digit {i}")
    ax.axis('off')

plt.tight_layout()
plt.savefig('figures/sample-digits.pdf')

```
